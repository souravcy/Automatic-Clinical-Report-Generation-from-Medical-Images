```python
import torch
from transformers import ViTFeatureExtractor, ViTModel
from PIL import Image

model_name = "google/vit-base-patch16-224-in21k"
feature_extractor = ViTFeatureExtractor.from_pretrained(model_name)
model = ViTModel.from_pretrained(model_name)

image_path = "/content/drive/MyDrive/CXR1_1_IM-0001-4001.png"
image = Image.open(image_path)

patch_size = 16
inputs = feature_extractor(images=image, return_tensors="pt", padding="max_length", max_length=1024)

with torch.no_grad():
    outputs = model(**inputs)

features = outputs.last_hidden_state
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np


model = ResNet50(weights = 'imagenet', include_top = False)

img_path = "/content/drive/MyDrive/CXR1_1_IM-0001-4001.png"

img = image.load_img(img_path , target_size = (224,224))

x = image.img_to_array(img)
x = np.expand_dims(x,axis = 0)
x = preprocess_input(x)

features = model.predict(x)
print(features)
```
